```python
import requests
import pandas as pd
import csv
from bs4 import BeautifulSoup as bs 
import json
import datetime
```

```python
url='https://www.worldathletics.org/records/by-category/world-records'
#Create a handle, page, to handle the contents of the website
page = requests.get(url)
#Store the contents of the website under doc
soup = bs(page.content)
```

```python
tables_out= soup.find("div", {"id": "womenoutdoor"})
tables_in= soup.find("div", {"id": "womenindoor"})

tableheader_out= tables_out.find_all('th')
tableheader_in= tables_in.find_all('tr')

table= soup.find_all('th')

print(table)
```

```python
data=[]
for tr in tables_out.find_all('tr'):    
    for th in tr.find_all('th'):
        new_data = th.text
        new_data = new_data.split()
        data.append(new_data[0])
print(data)
```

```python
disciplines_out = []
for td in tables_out.find_all('td', {'data-th':"DISCIPLINE"}):
    new_data = td.text
    new_data = new_data.split()
    separator = ' '
    new_data = separator.join(new_data)
    disciplines_out.append(new_data)

disciplines_out
```

```python
pref_out = []
for td in tables_out.find_all('td', {'data-th':"PERF"}):
    new_data = td.text
    new_data = new_data.split()
    separator = ''
    new_data = separator.join(new_data)
    if(new_data != ''):
        pref_out.append(new_data)
    
pref_out    
```

```python

for td in tables_out.find_all('td', {'data-th':"WIND"}):
    data.append(td.text)
    
```

```python
competitor_out = []
for td in tables_out.find_all('td', {'data-th':"COMPETITOR"}):
    new_data = td.text
    new_data = new_data.split()
    separator = ' '
    new_data = separator.join(new_data)
    competitor_out.append(new_data)
competitor_out
```

```python
DOB_out = []
for td in tables_out.find_all('td', {'data-th':"DOB"}):
    new_data = td.text
    new_data = new_data.split()
    separator = ' '
    new_data = separator.join(new_data)
    DOB_out.append(new_data)
DOB_out
```

```python
country_out = []
for td in tables_out.find_all('td', {'data-th':"COUNTRY"}):
    new_data = td.text
    new_data = new_data.split()
    separator = ' '
    new_data = separator.join(new_data)
    country_out.append(new_data)
country_out
```

```python
date_out = []
for td in tables_out.find_all('td', {'data-th':"DATE"}):
    new_data = td.text
    new_data = new_data.split()
    separator = ' '
    new_data = separator.join(new_data)
    date_out.append(new_data)
date_out
```

```python
data_df = pd.DataFrame({'Discipline': disciplines_out,
                        'Perf': pref_out,
                        'Competitor': competitor_out,
                        'DOB': DOB_out,
                        'Country': country_out,
                        'Date':date_out
                       })
for index, row in data_df.iterrows():
    row['Date']= row['Date'][7:11]

data_df = data_df.sort_values("Date")
data_df
```
