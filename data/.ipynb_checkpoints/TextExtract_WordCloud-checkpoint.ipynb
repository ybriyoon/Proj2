```python
import pandas as pd
path = "complete_data.csv"
data_df = pd.read_csv(path)
data_df
```

```python
desc_string = ""
```

```python
for desc in data_df["description"]:
    desc_string = f"{desc_string} {desc}"
desc_string = desc_string.replace('\n', '')
```

```python
desc_string
```

```python
text_file = open("description_string.txt", "w")
n = text_file.write(desc_string)
text_file.close()
```
